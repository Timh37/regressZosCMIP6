```python
import xarray as xr
import os
import matplotlib.pyplot as plt
```

Load the regression coefficient file:

```python
in_dir = '/Volumes/Naamloos/PhD_Data/CMIP6/regression/zoszostoga_zostoga'
coefs = xr.open_dataset(os.path.join(in_dir,'zoszostoga_zostoga_linregression_coefs.nc'))
```

Plot coefficients for given model for each SSP, for SSPs merged and for the mean of SSPs:

```python
model_to_plot = 'CanESM5'
```

Plot intercept:

```python
plt.figure() #plot coefficients for each SSP, including for all SSPs combined (ssp='merged')
coefs.sel(source_id=model_to_plot).polyfit_coefficients.sel(degree=0).plot(col="ssp", col_wrap=3,vmin=-1,vmax=1,cmap='seismic')

plt.figure() #plot SSP-averaged coefficients
coefs.sel(ssp=['ssp119','ssp126','ssp245','ssp370','ssp585']).sel(source_id=model_to_plot).polyfit_coefficients.sel(degree=0).mean(dim='ssp').plot(vmin=-1,vmax=1,cmap='seismic')
plt.title('mean of SSPs')
```

Plot degree-1:

```python
plt.figure() #plot coefficients for each SSP, including for all SSPs combined (ssp='merged')
coefs.sel(source_id=model_to_plot).polyfit_coefficients.sel(degree=1).plot(col="ssp", col_wrap=3,vmin=-3,vmax=3,cmap='seismic')
plt.figure() #plot SSP-averaged coefficients
coefs.sel(ssp=['ssp119','ssp126','ssp245','ssp370','ssp585']).sel(source_id=model_to_plot).polyfit_coefficients.sel(degree=1).mean(dim='ssp').plot(vmin=-3,vmax=3,cmap='seismic')
plt.title('mean of SSPs')
```
